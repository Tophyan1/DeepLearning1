```python
# PIL for installing images
from PIL import Image
import numpy as np
```

```python
# loading images
image1 = Image.open('icon1.jpg')
image2 = Image.open('icon2.jpg')
image3 = Image.open('icon3.jpg')
image4 = Image.open('icon4.jpg')
image5 = Image.open('icon5.jpg')
```

```python
# converting images to matrices
# convert int to float for floating point arithmetic later
array1 = np.asarray(image1, float)
array2 = np.asarray(image2, float)
array3 = np.asarray(image3, float)
array4 = np.asarray(image4, float)
array5 = np.asarray(image5, float)
arrays = [array1, array2, array3, array4, array5]
```

```python
# Exercise 1: printing the arrays
for i in arrays:
    print(i)
```

```python
colours = ["Red", "Green", "Blue"]
for array in arrays:
    # third axis is size 0, will concatenate to it
    finalArray = np.empty([128, 128, 0])
    for i in range(3): # 3 colour channels
        print(colours[i])
        # average (mean)
        average = np.average(array[:, :, i])
        # standard deviation
        stdDev = np.std(array[:, :, i])
        print("\tBefore:")
        print(f"\t\tAverage: \t\t{average}")
        print(f"\t\tStandard deviation: \t{stdDev}")
        stdArray = (array[:,:,i] - average) / stdDev #standardizing
        # concatenate to output array
        finalArray = np.dstack((finalArray, stdArray))
        # rounding to nearest integer because of floating point inaccuracies
        # (averages are in the 1E-18 range before rounding)
        # (standard deviations are either 1.0 or 0.99999... before rounding)
        print("\tAfter:")
        print(f"\t\tAverage: \t\t{np.average(finalArray[:,:,i]).round(0)}")
        print(f"\t\tStandard deviation: \t{np.std(finalArray[:,:,i]).round(0)}")
```

```python
# for dealing with .wav files
from scipy.io import wavfile
# first element is sample rate, second is the data, already numpy array
sampleRate, data = wavfile.read('priest_convert_wololo1.wav') # Wololo
# I don't know why the warning is there, but it seems to work just fine
```

```python
from scipy import signal
import matplotlib.pyplot as plt

# processing the signal
frequencies, times, spectrogram = signal.spectrogram(data, sampleRate)
# setting up the plot
plt.pcolormesh(times, frequencies, 10*np.log10(spectrogram), shading='gouraud')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

```python
# Python wrapper for the Wikipedia API
import wikipediaapi
# specifying the english language Wikipedia
wiki = wikipediaapi.Wikipedia('en')
# loading in article called 'Deep learning'
page = wiki.page('Deep learning')

print(page.text)
```

```python
from collections import Counter

pageText = page.text

# setting up histogram data
letterCounter = Counter(pageText.lower())
counts = letterCounter.values()
letters = letterCounter.keys()

#graphing histogram
barLocations = np.arange(len(counts))
plt.bar(barLocations, counts, align='center')
plt.xticks(barLocations, letters)
plt.show()
```
